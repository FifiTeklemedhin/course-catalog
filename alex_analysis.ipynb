```python
import numpy as np
import csv
import re
import datetime
```

```python
# Open the csv
csvfile = open('course_catalog_dump.csv','r')
# Turn the csv into an OrderedDict
courses=csv.DictReader(csvfile)
# Regex object to grab strings of four digits
time_regex = re.compile('\d{4}')
# Regex object to grab AM or PM
meridian_regex = re.compile('([AP]M)')
#week_regex = re.compile
for course in courses:
    time_string = time_regex.findall(course['schedule'])
    meridian_string = meridian_regex.findall(course['schedule'])
    timespan=[datetime.time(hour=0,minute=0),datetime.time(hour=0,minute=0)]
    if len(time_string) > 1:
        for i in range(2):
            hr = int(time_string[i][0:2])
            mn = int(time_string[i][2:4])
            mn = int(round(float(mn)/5)*5) % 60
            meridian = meridian_string[i]
            if meridian == 'AM' or hr == 12:
                timespan[i]=datetime.time(hour=hr,minute=mn)
            else:
                timespan[i]=datetime.time(hour=12+hr,minute=mn)
        print(course['number'])
        print(timespan)
    
        
        
```
